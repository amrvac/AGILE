```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbox_style=dict(boxstyle="round", fc='#ffffff', ec="w", alpha=0.75)
```

```python
# 1. Read the CSV file (update filename as needed)
fname = "profiles.csv"   # <- change to your actual filename

df = pd.read_csv(fname)

print(df.columns)

# 2. Plot rho vs "Points:1"
x = df["Points:0"]
rho = df["rho"]
p = df["p"]
v1 = df["v1"]/df["lfac"]

plt.figure(figsize=(6, 4))
plt.plot(x, rho/12, linestyle="-")
plt.plot(x, p/25, linestyle="--")
plt.plot(x, v1, linestyle=":")
plt.xlabel("x")
plt.title("Marti and Mueller (1994); Figure 6")
plt.grid(True, alpha=0.3)
plt.xlim(0,1)
plt.ylim(-1,1)

plt.annotate(r'$v$', (0.035, 0.16), xycoords='axes fraction', textcoords='axes fraction', bbox=bbox_style, ha="left", va="top")
plt.annotate(r'$\rho/12$', (0.035, 0.62), xycoords='axes fraction', textcoords='axes fraction', bbox=bbox_style, ha="left", va="top")
plt.annotate(r'$p/25$', (0.035, 0.78), xycoords='axes fraction', textcoords='axes fraction', bbox=bbox_style, ha="left", va="top")

plt.tight_layout()

# Save to PDF
plt.savefig("reference.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
